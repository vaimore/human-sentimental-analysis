```python
import matplotlib.pyplot as plt
from fer import FER

%matplotlib inline

test_image_one = plt.imread("E:/Vaibhav/Study_CS/Python/fear.jpeg")
```

```python
emo_detector = FER(mtcnn=True)
```

```python
captured_emotions = emo_detector.detect_emotions(test_image_one)
```

```python
print(captured_emotions)
```

```python
# Use the top Emotion() function to call for the dominant emotion in the image
dominant_emotion, emotion_score = emo_detector.top_emotion(test_image_one)
print(dominant_emotion, emotion_score)
```

```python
# Display the image
plt.imshow(test_image_one)

# Highlight the highest emotion score
highlight_text = f'Dominant Emotion: {dominant_emotion}\nEmotion Score: {emotion_score:.2f}'
plt.annotate(highlight_text, xy=(0.5, 0.5), xytext=(0.5, 0.8),
             arrowprops=dict(facecolor='red', shrink=0.05),
             fontsize=12, color='white',
             horizontalalignment='center', verticalalignment='center',
             bbox=dict(boxstyle='round,pad=0.3', edgecolor='red', facecolor='red'))

plt.show()
```
